## numpy는 행렬을 다루기 위해, pyplot은 그래프를 그리기 위해 import

```
import numpy as np
import matplotlib.pyplot as plt
```

## 1부터 6까지 자연수를 랜덤하게 뽑는다.

```
import random
print(random.randint(1, 6)) #1부터 6사이의 임의의 정수를 추출한다.
```

## 주사위 12번을 던지는 시뮬레이션을 5회 실시한다

```
n=12 #n번 던지는 시뮬레이션을
m=5  #m번 반복한다
eps = 0.05
count=0
for j in range(m) :
  dice = []
  X=[0,0,0,0,0,0]
  for i in range(n) :
    dice.append(random.randint(1,6))
  for k in range(6) :
    X[k]=dice.count(k+1)         #X[0]은 1이 몇 번 나왔는지, X[k]는 (k+1)이 몇 번 나왔는지
  if abs(X[2]/n-1/6)<eps :       #3이 몇 번 나왔는지
    count = count + 1            #P(|X/n-1/6|<0.05)인 경우가 몇 번인지 count로 출력한다
  print(j+1,"번째 결과 : ",dice) #의도한 대로 n번 던지는 시뮬레이션이 m번 반복되는지 확인한다
  plt.subplot(1,m,j+1)           #m개의 그래프를 그려준다
  plt.bar(np.arange(1,7),X)      #그래프 형태는 바 그래프
  plt.xlabel(j+1)                #몇번째 시행인지 표기
  plt.xticks(np.arange(1,7))     #각 눈금의 개수를 세어준다
  plt.axhline(y= n/6, color='red', linestyle='--', linewidth='1') #상대도수 = 1/6을 표시해준다
  plt.yticks([])                 #y 좌표를 깔끔하게
print("P(|X/",n,"-1/6|<",eps,") = ",count,"/",m) #m번 중 count번이 우리가 구하는 결과이다
```
